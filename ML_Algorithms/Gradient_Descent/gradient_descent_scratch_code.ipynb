```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Generating a random regression dataset
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
plt.scatter(X, y)
```

```python
## Lets apply OLS
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, y)
print("Intercept (b): ", reg.intercept_)
print("Slope (m): ", reg.coef_)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red')
```

```python
## Lets apply Gradient Descent assuming slope is constant m = 78.35
## Also, let assume the starting value of intecept, b = 0
y_pred = ((78.35 * X) + 0).reshape(4)
```

```python
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='green', label='Gradient Descent(b = 0)')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
print("Loss Slope: ", loss_slope)
```

```python
## Lets take learning rate as 0.1.
lr = 0.1
step_size = loss_slope * lr
print("Step Size: ", step_size)
```

```python
## Calculating new intecept value after one step of gradient descent
b = b - step_size
print("New Intercept (b): ", b)
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred1, color='green', label='Gradient Descent(b = {})'.format(b))
plt.plot(X, y_pred, color='blue', label='Gradient Descent(b = 0)')
plt.legend()
plt.show()
```

```python
## Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
print("Loss Slope: ", loss_slope)
```

```python
step_size = loss_slope * lr
print("Step Size: ", step_size)
```

```python
b = b - step_size
print("New Intercept (b): ", b)
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred2, color='green', label='Gradient Descent(b = {})'.format(b))
plt.plot(X, y_pred1, color='blue', label='Gradient Descent(b = {})'.format(20.927763408209216))
plt.plot(X, y_pred, color='cyan', label='Gradient Descent(b = 0)')
plt.plot()
plt.legend()
```

```python
## Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
print("Loss Slope: ", loss_slope)
```

```python
step_size = loss_slope * lr
print("Step Size: ", step_size)
```

```python
b = b - step_size
print("New Intercept (b): ", b)
```

```python
y_pred3 = ((78.35 * X) + b).reshape(4)
plt.scatter(X, y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred3, color='green', label='Gradient Descent(b = {})'.format(b))
plt.plot(X, y_pred2, color='blue', label='Gradient Descent(b = {})'.format(25.11331608985106))
plt.plot(X, y_pred1, color='cyan', label='Gradient Descent(b = {})'.format(20.927763408209216))
plt.plot(X, y_pred, color='magenta', label='Gradient Descent(b = 0)')
plt.plot()
```

```python
## Lets take b = -100 and run it in loop
b = -100
m = 78.35
lr = 0.1
epochs = 10
for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    b = b - (lr * loss_slope)
    y_pred = m * X + b
    plt.plot(X, y_pred)
    print("Epoch: {}, Intercept (b): {}".format(i+1, b))
plt.scatter(X, y)
```
